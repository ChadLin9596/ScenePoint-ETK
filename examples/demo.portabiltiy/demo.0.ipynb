```python
import os
import scene_point_etk.scene_db as scene_db
import py_utils.utils
```

```python
new_scenes_root = "/home/username/Data/a_new_path"
```

---
export to portable version

```python
scene_ids = scene_db.list_scene_ids()

prog = py_utils.utils.ProgressTimer()
prog.tic(len(scene_ids))

for scene_id in scene_ids:
    scene = scene_db.OriginalScene(scene_id)
    scene.export(
        new_scenes_root,
        skip_scene=False,
        skip_cameras=True,
    )
    prog.toc()
```

```python
scene_pairs = scene_db.list_scene_version_pairs()

prog = py_utils.utils.ProgressTimer()
prog.tic(len(scene_pairs))

for scene_id, scene_version in scene_pairs:
    scene = scene_db.EditedScene(scene_id, scene_version)
    scene.export(
        new_scenes_root,
        skip_scene=False,
        skip_cameras=True,
    )
    prog.toc()
```

volumn comparsion

```python
# source directory


def func(x):
    return os.path.isfile(x)


files = py_utils.utils.find_directory_by_DFS(
    root=scene_db.SCENE_ROOT,
    condition=func,
)

num_bytes = sum([os.path.getsize(i) for i in files])
num_gigabytes = num_bytes / 1024 / 1024 / 1024
num_gigabytes
```

```python
# destination directory


def func(x):
    return os.path.isfile(x)


files = py_utils.utils.find_directory_by_DFS(
    root=new_scenes_root,
    condition=func,
)

num_bytes = sum([os.path.getsize(i) for i in files])
num_gigabytes = num_bytes / 1024 / 1024 / 1024
num_gigabytes
```
